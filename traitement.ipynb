```python
import numpy as np
import pandas as pd
from tensorflow.keras.utils import load_img
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization
import random
import os
import cv2
import seaborn as sns
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras import models, layers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import warnings
warnings.filterwarnings('ignore')

DATA_DIR = 'cell_images'
CATEGORIES = ['Parasitized', 'Uninfected']
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 20
RANDOM_STATE = 42
MODELS_DIR = 'models'
os.makedirs(MODELS_DIR, exist_ok=True)

# Fixer toutes les graines al√©atoires pour la reproductibilit√©
random.seed(RANDOM_STATE)
np.random.seed(RANDOM_STATE)
import tensorflow as tf

tf.random.set_seed(RANDOM_STATE)

# 1) CHARGEMENT DE L'ENSEMBLE DE DONN√âES
print("=" * 70)
print("1) CHARGEMENT DE L'ENSEMBLE DE DONN√âES")
print("=" * 70)

filepaths = []
labels = []
label_indices = []  # Pour utiliser to_categorical plus tard

for idx, cat in enumerate(CATEGORIES):
    folder = os.path.join(DATA_DIR, cat)
    if not os.path.isdir(folder):
        raise FileNotFoundError(f"Dossier attendu non trouv√©: {folder}")

    files = [f for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    print(f"  - {cat}: {len(files)} images trouv√©es")

    for fname in files:
        filepaths.append(os.path.join(folder, fname))
        labels.append(cat)
        label_indices.append(idx)

# Utilisation de pandas.DataFrame
df = pd.DataFrame({
    'filepath': filepaths,
    'label': labels,
    'label_index': label_indices
})

# Utilisation de random via pandas.sample
df = df.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

print(f"\nTotal images charg√©es: {len(df)}")
print("\nDistribution des classes:")
print(df['label'].value_counts())

# 2) PR√âTRAITEMENT ‚Äî V√©rification et nettoyage des images
print("\n" + "=" * 70)
print("2) PR√âTRAITEMENT ET V√âRIFICATION DES IMAGES")
print("=" * 70)


def preprocess_and_verify_images(df, img_size=IMG_SIZE):
    """
    Utilise cv2, load_img et img_to_array pour v√©rifier les images.
    Utilise numpy pour les calculs statistiques.
    """
    valid_paths = []
    valid_labels = []
    valid_indices = []
    corrupted = []
    img_shapes = []
    pixel_means = []

    for idx, row in df.iterrows():
        try:
            # Test 1: Chargement avec OpenCV (cv2)
            img_cv = cv2.imread(row['filepath'])
            if img_cv is None:
                raise ValueError("Image non chargeable avec cv2")

            # Conversion BGR -> RGB avec cv2
            img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

            # V√©rification des dimensions avec numpy
            if img_cv.shape[0] < 10 or img_cv.shape[1] < 10:
                raise ValueError("Image trop petite")

            # Test 2: Redimensionnement avec cv2
            img_resized = cv2.resize(img_cv_rgb, img_size)

            # Test 3: Chargement avec Keras (load_img)
            img_keras = load_img(row['filepath'], target_size=img_size)

            # Test 4: Conversion en array avec img_to_array
            img_array = img_to_array(img_keras)

            # Calcul statistique avec numpy
            mean_val = np.mean(img_array)

            # Si tout est OK, on conserve l'image
            valid_paths.append(row['filepath'])
            valid_labels.append(row['label'])
            valid_indices.append(row['label_index'])
            img_shapes.append(img_cv.shape)
            pixel_means.append(mean_val)

        except Exception as e:
            corrupted.append({'filepath': row['filepath'], 'error': str(e)})
            print(f"  ‚ö† Image corrompue: {os.path.basename(row['filepath'])} - {e}")

    # Cr√©er DataFrame avec pandas
    clean_df = pd.DataFrame({
        'filepath': valid_paths,
        'label': valid_labels,
        'label_index': valid_indices
    })

    # Statistiques avec numpy
    if img_shapes:
        heights = np.array([s[0] for s in img_shapes])
        widths = np.array([s[1] for s in img_shapes])
        means = np.array(pixel_means)

        print(f"\n‚úì Images valides: {len(clean_df)}/{len(df)}")
        print(f"‚úì Images corrompues retir√©es: {len(corrupted)}")
        print(f"\nStatistiques (numpy):")
        print(f"  Hauteur - min: {np.min(heights)}, max: {np.max(heights)}, moyenne: {np.mean(heights):.1f}")
        print(f"  Largeur - min: {np.min(widths)}, max: {np.max(widths)}, moyenne: {np.mean(widths):.1f}")
        print(
            f"  Valeur moyenne pixels - min: {np.min(means):.1f}, max: {np.max(means):.1f}, moyenne: {np.mean(means):.1f}")

    return clean_df, corrupted


# Appliquer le pr√©traitement
df_clean, corrupted_images = preprocess_and_verify_images(df)
df = df_clean.reset_index(drop=True)

# Sauvegarder avec pandas
if corrupted_images:
    corrupted_df = pd.DataFrame(corrupted_images)
    corrupted_df.to_csv(os.path.join(MODELS_DIR, 'corrupted_images.csv'), index=False)
    print(f"\n‚úì Liste des images corrompues sauvegard√©e")

# 3) AFFICHER DES EXEMPLES PAR CAT√âGORIE
print("\n" + "=" * 70)
print("3) AFFICHAGE D'EXEMPLES PAR CAT√âGORIE")
print("=" * 70)

# Utilisation de matplotlib.pyplot et seaborn
plt.style.use('seaborn-v0_8-darkgrid')
fig = plt.figure(figsize=(14, 8))

for i, cat in enumerate(CATEGORIES):
    # Utilisation de pandas pour filtrer et √©chantillonner
    cat_df = df[df['label'] == cat]
    samples = cat_df.sample(min(4, len(cat_df)), random_state=RANDOM_STATE)

    for j, (idx, row) in enumerate(samples.iterrows()):
        # M√©thode 1: Chargement avec cv2
        img_cv = cv2.imread(row['filepath'])
        img_cv = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
        img_cv = cv2.resize(img_cv, IMG_SIZE)

        # Utilisation de matplotlib.pyplot
        plt.subplot(2, 4, i * 4 + j + 1)
        plt.imshow(img_cv)
        plt.title(f"{cat}\nExemple {j + 1}", fontsize=10)
        plt.axis('off')

plt.suptitle('√âchantillons d\'images par cat√©gorie', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'sample_images.png'), dpi=150, bbox_inches='tight')
plt.show()

# Affichage suppl√©mentaire avec seaborn
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# Utilisation de seaborn pour visualisation
class_counts = df['label'].value_counts()
sns.barplot(x=class_counts.index, y=class_counts.values, ax=axes[0], palette='viridis')
axes[0].set_title('Distribution des classes', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Nombre d\'images')
axes[0].set_xlabel('Classe')

# Pie chart avec matplotlib
axes[1].pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%',
            colors=sns.color_palette('viridis', len(class_counts)))
axes[1].set_title('Proportion des classes', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'class_distribution.png'), dpi=150, bbox_inches='tight')
plt.show()

# 4) NORMALISATION DES DONN√âES D'IMAGE
print("\n" + "=" * 70)
print("4) ANALYSE ET NORMALISATION DES DONN√âES")
print("=" * 70)


def analyze_image_statistics(df, n_samples=100):
    """
    Utilise load_img, img_to_array, numpy, matplotlib et seaborn
    """
    # √âchantillonnage avec pandas
    sample_df = df.sample(min(n_samples, len(df)), random_state=RANDOM_STATE)
    means = []
    stds = []
    mins = []
    maxs = []

    print(f"Analyse de {len(sample_df)} images √©chantillons...")

    for _, row in sample_df.iterrows():
        # Chargement avec load_img
        img = load_img(row['filepath'], target_size=IMG_SIZE)
        # Conversion avec img_to_array
        img_array = img_to_array(img)

        # Calculs statistiques avec numpy
        means.append(np.mean(img_array))
        stds.append(np.std(img_array))
        mins.append(np.min(img_array))
        maxs.append(np.max(img_array))

    # Conversion en arrays numpy
    means = np.array(means)
    stds = np.array(stds)
    mins = np.array(mins)
    maxs = np.array(maxs)

    print(f"\nStatistiques calcul√©es avec numpy:")
    print(f"  Valeur moyenne des pixels: {np.mean(means):.2f} ¬± {np.std(means):.2f}")
    print(f"  √âcart-type des pixels: {np.mean(stds):.2f} ¬± {np.std(stds):.2f}")
    print(f"  Min global: {np.min(mins):.2f}, Max global: {np.max(maxs):.2f}")
    print(f"  Range apr√®s normalisation: [0.0, 1.0]")

    # Visualisation avec matplotlib et seaborn
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # Histogramme des moyennes avec seaborn
    sns.histplot(means, bins=30, kde=True, ax=axes[0, 0], color='blue', edgecolor='black')
    axes[0, 0].set_title('Distribution des moyennes des pixels', fontsize=11, fontweight='bold')
    axes[0, 0].set_xlabel('Valeur moyenne')
    axes[0, 0].set_ylabel('Fr√©quence')
    axes[0, 0].axvline(np.mean(means), color='red', linestyle='--', label=f'Moyenne: {np.mean(means):.1f}')
    axes[0, 0].legend()

    # Histogramme des √©carts-types avec seaborn
    sns.histplot(stds, bins=30, kde=True, ax=axes[0, 1], color='orange', edgecolor='black')
    axes[0, 1].set_title('Distribution des √©carts-types', fontsize=11, fontweight='bold')
    axes[0, 1].set_xlabel('√âcart-type')
    axes[0, 1].set_ylabel('Fr√©quence')
    axes[0, 1].axvline(np.mean(stds), color='red', linestyle='--', label=f'Moyenne: {np.mean(stds):.1f}')
    axes[0, 1].legend()

    # Boxplot avec seaborn
    data_for_box = pd.DataFrame({'Moyennes': means, '√âcarts-types': stds})
    sns.boxplot(data=data_for_box, ax=axes[1, 0], palette='Set2')
    axes[1, 0].set_title('Boxplot des statistiques', fontsize=11, fontweight='bold')
    axes[1, 0].set_ylabel('Valeur')

    # Scatter plot avec matplotlib
    axes[1, 1].scatter(means, stds, alpha=0.6, c=means, cmap='viridis', edgecolors='black')
    axes[1, 1].set_title('Relation Moyenne vs √âcart-type', fontsize=11, fontweight='bold')
    axes[1, 1].set_xlabel('Moyenne des pixels')
    axes[1, 1].set_ylabel('√âcart-type des pixels')
    axes[1, 1].colorbar = plt.colorbar(axes[1, 1].collections[0], ax=axes[1, 1])

    plt.tight_layout()
    plt.savefig(os.path.join(MODELS_DIR, 'pixel_statistics.png'), dpi=150, bbox_inches='tight')
    plt.show()

    return np.mean(means), np.mean(stds)


# Analyser les statistiques
mean_pixel, std_pixel = analyze_image_statistics(df, n_samples=200)

print("\nüìä Normalisation:")
print("  ‚Üí M√©thode: Division par 255 (rescale=1./255)")
print("  ‚Üí Transformation: [0, 255] ‚Üí [0.0, 1.0]")

# 5) DIVISION DES DONN√âES - Utilisation de sklearn.train_test_split
print("\n" + "=" * 70)
print("5) DIVISION DES DONN√âES (train/val/test)")
print("=" * 70)

# Stratified split avec sklearn
X = df['filepath'].values  # numpy array
y = df['label_index'].values  # numpy array

# Premier split: train+val vs test (sklearn.train_test_split)
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, y,
    test_size=0.15,
    stratify=y,
    random_state=RANDOM_STATE
)

# Deuxi√®me split: train vs val (sklearn.train_test_split)
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval, y_trainval,
    test_size=0.15,  # 15% de trainval pour validation
    stratify=y_trainval,
    random_state=RANDOM_STATE
)

print(f"Tailles des ensembles:")
print(f"  - Train: {len(X_train)} images ({len(X_train) / len(X) * 100:.1f}%)")
print(f"  - Validation: {len(X_val)} images ({len(X_val) / len(X) * 100:.1f}%)")
print(f"  - Test: {len(X_test)} images ({len(X_test) / len(X) * 100:.1f}%)")

# Conversion avec to_categorical (keras.utils)
y_train_cat = to_categorical(y_train, num_classes=len(CATEGORIES))
y_val_cat = to_categorical(y_val, num_classes=len(CATEGORIES))
y_test_cat = to_categorical(y_test, num_classes=len(CATEGORIES))

print(f"\nFormes apr√®s to_categorical:")
print(f"  y_train_cat: {y_train_cat.shape}")
print(f"  y_val_cat: {y_val_cat.shape}")
print(f"  y_test_cat: {y_test_cat.shape}")


# Chargement et normalisation des images
def load_and_normalize_images(filepaths, target_size=IMG_SIZE):
    """
    Utilise load_img, img_to_array et numpy pour charger et normaliser
    """
    images = []
    for path in filepaths:
        # Chargement avec load_img
        img = load_img(path, target_size=target_size)
        # Conversion avec img_to_array
        img_array = img_to_array(img)
        # Normalisation avec numpy
        img_normalized = img_array / 255.0
        images.append(img_normalized)

    # Conversion en numpy array
    return np.array(images)


print("\nChargement des images en m√©moire...")
X_train_images = load_and_normalize_images(X_train)
X_val_images = load_and_normalize_images(X_val)
X_test_images = load_and_normalize_images(X_test)

print(f"\nFormes des donn√©es:")
print(f"  X_train: {X_train_images.shape}")
print(f"  X_val: {X_val_images.shape}")
print(f"  X_test: {X_test_images.shape}")

# V√©rification de la normalisation avec numpy
print(f"\nV√©rification normalisation (numpy):")
print(f"  Min: {np.min(X_train_images):.4f}")
print(f"  Max: {np.max(X_train_images):.4f}")
print(f"  Mean: {np.mean(X_train_images):.4f}")

# 6) CR√âATION DES MOD√àLES CNN - Utilisation de keras.models et keras.layers
print("\n" + "=" * 70)
print("6) CR√âATION DES 3 MOD√àLES CNN")
print("=" * 70)


def build_model_A(input_shape=(*IMG_SIZE, 3), n_classes=2):
    """Mod√®le A - Architecture simple"""
    # Utilisation de Sequential de keras.models
    model = Sequential([
        # Utilisation de Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense
        Conv2D(32, (3, 3), padding='same', input_shape=input_shape),
        Activation('relu'),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3)),
        Activation('relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128),
        Activation('relu'),
        Dropout(0.5),
        Dense(n_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("  ‚úì Mod√®le A cr√©√© (architecture simple)")
    return model


def build_model_B(input_shape=(*IMG_SIZE, 3), n_classes=2):
    """Mod√®le B - Architecture avec BatchNormalization"""
    model = Sequential()
    # Ajout de BatchNormalization
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=input_shape))
    model.add(Conv2D(32, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(n_classes, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("  ‚úì Mod√®le B cr√©√© (avec BatchNormalization)")
    return model


def build_model_C(input_shape=(*IMG_SIZE, 3), n_classes=2):
    """Mod√®le C - Architecture profonde"""
    model = Sequential()
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=input_shape))
    model.add(Conv2D(32, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(n_classes, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    print("  ‚úì Mod√®le C cr√©√© (architecture profonde)")
    return model


# Cr√©ation des mod√®les
models_dict = {
    'model_A': build_model_A(),
    'model_B': build_model_B(),
    'model_C': build_model_C()
}

# 7) ENTRA√éNEMENT - Utilisation de callbacks (EarlyStopping, etc.)
print("\n" + "=" * 70)
print("7) ENTRA√éNEMENT DES MOD√àLES")
print("=" * 70)

histories = {}
metrics = []

for name, model in models_dict.items():
    print(f"\n{'=' * 60}")
    print(f"Entra√Ænement de {name}")
    print(f"{'=' * 60}")

    # Callbacks: EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
    ckpt_path = os.path.join(MODELS_DIR, f"{name}_best.h5")
    checkpoint = ModelCheckpoint(ckpt_path, monitor='val_accuracy',
                                 save_best_only=True, mode='max', verbose=1)
    early = EarlyStopping(monitor='val_loss', patience=6,
                          restore_best_weights=True, verbose=1)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                                  patience=3, min_lr=1e-6, verbose=1)

    # Entra√Ænement
    history = model.fit(
        X_train_images, y_train_cat,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=(X_val_images, y_val_cat),
        callbacks=[checkpoint, early, reduce_lr],
        verbose=2
    )

    histories[name] = history

    # Calcul des m√©triques avec numpy
    best_val_acc = np.max(history.history['val_accuracy'])
    best_val_loss = np.min(history.history['val_loss'])

    metrics.append({
        'model': name,
        'val_accuracy': float(best_val_acc),
        'val_loss': float(best_val_loss)
    })

# 8) √âVALUATION - Utilisation de matplotlib, seaborn, pandas
print("\n" + "=" * 70)
print("8) √âVALUATION ET VISUALISATION")
print("=" * 70)

# Cr√©ation DataFrame avec pandas
metrics_df = pd.DataFrame(metrics).sort_values('val_accuracy', ascending=False)
metrics_df.to_csv(os.path.join(MODELS_DIR, 'metrics_comparison.csv'), index=False)

print("\nüìä Comparaison des mod√®les:")
print(metrics_df.to_string(index=False))

# Visualisation avec matplotlib et seaborn
fig, axes = plt.subplots(len(models_dict), 2, figsize=(14, 4 * len(models_dict)))

for idx, (name, history) in enumerate(histories.items()):
    # Graphique Accuracy
    axes[idx, 0].plot(history.history['accuracy'], label='Train', linewidth=2)
    axes[idx, 0].plot(history.history['val_accuracy'], label='Validation', linewidth=2)
    axes[idx, 0].set_title(f'Accuracy - {name}', fontsize=12, fontweight='bold')
    axes[idx, 0].set_xlabel('Epoch')
    axes[idx, 0].set_ylabel('Accuracy')
    axes[idx, 0].legend()
    axes[idx, 0].grid(True, alpha=0.3)

    # Graphique Loss
    axes[idx, 1].plot(history.history['loss'], label='Train', linewidth=2)
    axes[idx, 1].plot(history.history['val_loss'], label='Validation', linewidth=2)
    axes[idx, 1].set_title(f'Loss - {name}', fontsize=12, fontweight='bold')
    axes[idx, 1].set_xlabel('Epoch')
    axes[idx, 1].set_ylabel('Loss')
    axes[idx, 1].legend()
    axes[idx, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'all_training_curves.png'), dpi=150, bbox_inches='tight')
plt.show()

# Comparaison visuelle avec seaborn
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barplot des accuracies avec seaborn
sns.barplot(data=metrics_df, x='model', y='val_accuracy', ax=axes[0], palette='viridis')
axes[0].set_title('Comparaison des Accuracies de Validation', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Validation Accuracy')
axes[0].set_ylim([0, 1])
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.4f')

# Barplot des losses avec seaborn
sns.barplot(data=metrics_df, x='model', y='val_loss', ax=axes[1], palette='rocket')
axes[1].set_title('Comparaison des Losses de Validation', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Validation Loss')
for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.4f')

plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'models_comparison.png'), dpi=150, bbox_inches='tight')
plt.show()

# 9) COMPARAISON FINALE
print("\n" + "=" * 70)
print("9) COMPARAISON DES MOD√àLES")
print("=" * 70)
print(metrics_df.to_string(index=False))

# 10) SAUVEGARDE DU MEILLEUR MOD√àLE - Utilisation de models.load_model
print("\n" + "=" * 70)
print("10) SAUVEGARDE ET TEST FINAL")
print("=" * 70)

best_row = metrics_df.iloc[0]
best_model_name = best_row['model']
best_model_file = os.path.join(MODELS_DIR, f"{best_model_name}_best.h5")

# Chargement avec tensorflow.keras.models
best_model = models.load_model(best_model_file)
best_overall_path = os.path.join(MODELS_DIR, 'best_overall_model.h5')
best_model.save(best_overall_path)

print(f"‚úì Meilleur mod√®le: {best_model_name}")
print(f"‚úì Sauvegard√© dans: {best_overall_path}")

# √âvaluation finale sur test - Utilisation de sklearn.metrics
print("\nüìä √âvaluation sur l'ensemble de test:")
y_pred_proba = best_model.predict(X_test_images, verbose=0)
y_pred = np.argmax(y_pred_proba, axis=1)  # numpy.argmax

# Utilisation de accuracy_score de sklearn
test_accuracy = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {test_accuracy:.4f} ({test_accuracy * 100:.2f}%)")

# Classification report avec sklearn
print("\nRapport de classification:")
print(classification_report(y_test, y_pred, target_names=CATEGORIES))

# Confusion Matrix - Utilisation de sklearn.metrics.confusion_matrix et seaborn
cm = confusion_matrix(y_test, y_pred)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Heatmap avec seaborn
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=CATEGORIES, yticklabels=CATEGORIES,
            cbar_kws={'label': 'Count'})
axes[0].set_title(f'Confusion Matrix - {best_model_name}', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Predicted Label')
axes[0].set_ylabel('True Label')

# Heatmap normalis√©e avec seaborn et numpy
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
sns.heatmap(cm_normalized, annot=True, fmt='.2%', cmap='Greens', ax=axes[1],
            xticklabels=CATEGORIES, yticklabels=CATEGORIES,
            cbar_kws={'label': 'Proportion'})
axes[1].set_title(f'Confusion Matrix (Normalized) - {best_model_name}', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Predicted Label')
axes[1].set_ylabel('True Label')

plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'confusion_matrix_best.png'), dpi=150, bbox_inches='tight')
plt.show()

# Calculs suppl√©mentaires avec numpy
TP = cm[1, 1]  # True Positives (Parasitized correctement class√©)
TN = cm[0, 0]  # True Negatives (Uninfected correctement class√©)
FP = cm[0, 1]  # False Positives (Uninfected class√© comme Parasitized)
FN = cm[1, 0]  # False Negatives (Parasitized class√© comme Uninfected)

# Calculs avec numpy
sensitivity = TP / (TP + FN) if (TP + FN) > 0 else 0
specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity) if (precision + sensitivity) > 0 else 0

print("\nüìà M√©triques d√©taill√©es (calcul√©es avec numpy):")
print(f"  Sensitivity (Recall): {sensitivity:.4f}")
print(f"  Specificity: {specificity:.4f}")
print(f"  Precision: {precision:.4f}")
print(f"  F1-Score: {f1_score:.4f}")

# Visualisation des m√©triques avec matplotlib et seaborn
metrics_names = ['Accuracy', 'Sensitivity', 'Specificity', 'Precision', 'F1-Score']
metrics_values = [test_accuracy, sensitivity, specificity, precision, f1_score]

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(metrics_names, metrics_values, color=sns.color_palette('viridis', len(metrics_names)),
              edgecolor='black', linewidth=1.5)
ax.set_ylim([0, 1.1])
ax.set_ylabel('Score', fontsize=12)
ax.set_title(f'M√©triques de Performance - {best_model_name} (Test Set)',
             fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)

# Ajouter les valeurs sur les barres
for bar, value in zip(bars, metrics_values):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2., height,
            f'{value:.3f}',
            ha='center', va='bottom', fontweight='bold', fontsize=11)

plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'performance_metrics.png'), dpi=150, bbox_inches='tight')
plt.show()

# Analyse des pr√©dictions - exemples corrects et incorrects
print("\n" + "=" * 70)
print("ANALYSE DES PR√âDICTIONS")
print("=" * 70)

# Identifier les pr√©dictions correctes et incorrectes avec numpy
correct_indices = np.where(y_pred == y_test)[0]
incorrect_indices = np.where(y_pred != y_test)[0]

print(
    f"\nPr√©dictions correctes: {len(correct_indices)}/{len(y_test)} ({len(correct_indices) / len(y_test) * 100:.2f}%)")
print(
    f"Pr√©dictions incorretes: {len(incorrect_indices)}/{len(y_test)} ({len(incorrect_indices) / len(y_test) * 100:.2f}%)")

# Afficher quelques exemples de pr√©dictions incorrectes
if len(incorrect_indices) > 0:
    n_show = min(8, len(incorrect_indices))
    # Utilisation de random.sample pour s√©lection al√©atoire
    random.seed(RANDOM_STATE)
    sample_incorrect = random.sample(list(incorrect_indices), n_show)

    fig = plt.figure(figsize=(16, 8))
    for i, idx in enumerate(sample_incorrect):
        # Chargement avec cv2
        img = cv2.imread(X_test[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, IMG_SIZE)

        true_label = CATEGORIES[y_test[idx]]
        pred_label = CATEGORIES[y_pred[idx]]
        confidence = y_pred_proba[idx][y_pred[idx]]

        plt.subplot(2, 4, i + 1)
        plt.imshow(img)
        plt.title(f'Vrai: {true_label}\nPr√©dit: {pred_label}\nConf: {confidence:.2%}',
                  fontsize=9, color='red')
        plt.axis('off')

    plt.suptitle('Exemples de Pr√©dictions Incorrectes', fontsize=14, fontweight='bold', color='red')
    plt.tight_layout()
    plt.savefig(os.path.join(MODELS_DIR, 'incorrect_predictions.png'), dpi=150, bbox_inches='tight')
    plt.show()

# Afficher quelques exemples de pr√©dictions correctes
if len(correct_indices) > 0:
    n_show = min(8, len(correct_indices))
    # Utilisation de random.sample
    sample_correct = random.sample(list(correct_indices), n_show)

    fig = plt.figure(figsize=(16, 8))
    for i, idx in enumerate(sample_correct):
        # Chargement avec cv2
        img = cv2.imread(X_test[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, IMG_SIZE)

        true_label = CATEGORIES[y_test[idx]]
        confidence = y_pred_proba[idx][y_pred[idx]]

        plt.subplot(2, 4, i + 1)
        plt.imshow(img)
        plt.title(f'Classe: {true_label}\nConfiance: {confidence:.2%}',
                  fontsize=9, color='green')
        plt.axis('off')

    plt.suptitle('Exemples de Pr√©dictions Correctes', fontsize=14, fontweight='bold', color='green')
    plt.tight_layout()
    plt.savefig(os.path.join(MODELS_DIR, 'correct_predictions.png'), dpi=150, bbox_inches='tight')
    plt.show()

# Distribution des confiances avec seaborn et numpy
print("\n" + "=" * 70)
print("ANALYSE DES CONFIANCES DE PR√âDICTION")
print("=" * 70)

# Extraire les confiances avec numpy
confidences_correct = y_pred_proba[correct_indices, y_pred[correct_indices]]
confidences_incorrect = y_pred_proba[incorrect_indices, y_pred[incorrect_indices]]

print(f"\nConfiance moyenne (pr√©dictions correctes): {np.mean(confidences_correct):.4f}")
print(f"Confiance moyenne (pr√©dictions incorrectes): {np.mean(confidences_incorrect):.4f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogramme avec seaborn
sns.histplot(confidences_correct, bins=30, kde=True, color='green',
             label='Correctes', ax=axes[0], alpha=0.6)
sns.histplot(confidences_incorrect, bins=30, kde=True, color='red',
             label='Incorrectes', ax=axes[0], alpha=0.6)
axes[0].set_xlabel('Confiance de pr√©diction')
axes[0].set_ylabel('Fr√©quence')
axes[0].set_title('Distribution des Confiances', fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Boxplot avec seaborn
conf_data = pd.DataFrame({
    'Confiance': np.concatenate([confidences_correct, confidences_incorrect]),
    'Type': ['Correctes'] * len(confidences_correct) + ['Incorrectes'] * len(confidences_incorrect)
})
sns.boxplot(data=conf_data, x='Type', y='Confiance', ax=axes[1], palette=['green', 'red'])
axes[1].set_title('Comparaison des Confiances', fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'confidence_analysis.png'), dpi=150, bbox_inches='tight')
plt.show()

# Matrice de corr√©lation des pr√©dictions entre mod√®les avec pandas et seaborn
print("\n" + "=" * 70)
print("CORR√âLATION DES PR√âDICTIONS ENTRE MOD√àLES")
print("=" * 70)

predictions_all = {}
for name, model in models_dict.items():
    # Recharger le meilleur mod√®le sauvegard√©
    model_path = os.path.join(MODELS_DIR, f"{name}_best.h5")
    loaded_model = models.load_model(model_path)
    preds = loaded_model.predict(X_test_images, verbose=0)
    predictions_all[name] = np.argmax(preds, axis=1)

# Cr√©er DataFrame avec pandas
predictions_df = pd.DataFrame(predictions_all)
predictions_df['True_Label'] = y_test

# Calculer la matrice de corr√©lation avec pandas
corr_matrix = predictions_df.corr()

# Visualiser avec seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='coolwarm',
            center=0, square=True, linewidths=1, cbar_kws={'label': 'Corr√©lation'})
plt.title('Matrice de Corr√©lation des Pr√©dictions entre Mod√®les',
          fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(MODELS_DIR, 'models_correlation.png'), dpi=150, bbox_inches='tight')
plt.show()

# Accord entre mod√®les avec numpy
print("\nAccord entre les mod√®les:")
for i, name1 in enumerate(predictions_all.keys()):
    for name2 in list(predictions_all.keys())[i + 1:]:
        agreement = np.mean(predictions_all[name1] == predictions_all[name2])
        print(f"  {name1} vs {name2}: {agreement:.2%}")

# R√âSUM√â FINAL
print("\n" + "=" * 70)
print("R√âSUM√â FINAL DU PIPELINE")
print("=" * 70)

summary = f"""
‚úì BIBLIOTH√àQUES UTILIS√âES:
  - numpy: {len([x for x in dir(np) if not x.startswith('_')])} fonctions disponibles
  - pandas: DataFrames pour gestion des donn√©es
  - cv2 (OpenCV): Chargement et traitement d'images
  - tensorflow.keras: load_img, img_to_array, to_categorical, models, layers, callbacks
  - keras: Sequential, Conv2D, MaxPooling2D, Dense, Dropout, Flatten, Activation, BatchNormalization
  - sklearn: train_test_split, confusion_matrix, accuracy_score, classification_report
  - matplotlib.pyplot: Visualisations graphiques
  - seaborn: Visualisations statistiques avanc√©es
  - random: Gestion de l'al√©atoire
  - os: Gestion des fichiers et dossiers

‚úì DONN√âES:
  - Total images: {len(df)}
  - Images corrompues retir√©es: {len(corrupted_images)}
  - Train: {len(X_train)} ({len(X_train) / len(df) * 100:.1f}%)
  - Validation: {len(X_val)} ({len(X_val) / len(df) * 100:.1f}%)
  - Test: {len(X_test)} ({len(X_test) / len(df) * 100:.1f}%)

‚úì MOD√àLES ENTRA√éN√âS: {len(models_dict)}
  - {', '.join(models_dict.keys())}

‚úì MEILLEUR MOD√àLE: {best_model_name}
  - Validation Accuracy: {best_row['val_accuracy']:.4f}
  - Test Accuracy: {test_accuracy:.4f}
  - Sensitivity: {sensitivity:.4f}
  - Specificity: {specificity:.4f}
  - F1-Score: {f1_score:.4f}

‚úì FICHIERS G√âN√âR√âS:
  - Mod√®les sauvegard√©s: {len(models_dict) + 1}
  - Graphiques g√©n√©r√©s: 9+
  - CSV de m√©triques: ‚úì
  - Dossier de sortie: {MODELS_DIR}/

‚úì NORMALISATION:
  - M√©thode: Division par 255
  - Range: [0.0, 1.0]
  - V√©rifi√©e avec numpy

‚úì CALLBACKS UTILIS√âS:
  - EarlyStopping: Arr√™t anticip√© si pas d'am√©lioration
  - ReduceLROnPlateau: R√©duction du learning rate
  - ModelCheckpoint: Sauvegarde du meilleur mod√®le
"""

print(summary)

# Sauvegarder le r√©sum√© avec pandas
summary_data = {
    'Metric': ['Total Images', 'Train Size', 'Val Size', 'Test Size',
               'Best Model', 'Test Accuracy', 'Sensitivity', 'Specificity',
               'Precision', 'F1-Score'],
    'Value': [len(df), len(X_train), len(X_val), len(X_test),
              best_model_name, f"{test_accuracy:.4f}", f"{sensitivity:.4f}",
              f"{specificity:.4f}", f"{precision:.4f}", f"{f1_score:.4f}"]
}
summary_df = pd.DataFrame(summary_data)
summary_df.to_csv(os.path.join(MODELS_DIR, 'final_summary.csv'), index=False)

print(f"\n‚úì R√©sum√© sauvegard√© dans: {os.path.join(MODELS_DIR, 'final_summary.csv')}")
print(f"\n{'=' * 70}")
print("PIPELINE TERMIN√â AVEC SUCC√àS! üéâ")
print(f"{'=' * 70}\n")
```
